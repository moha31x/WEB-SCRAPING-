```python
url = requests.get('https://www.premierleague.com/tables').text

soup = BeautifulSoup(url, 'lxml').body

table = soup.find('div', class_='tableContainer').table.find('tbody', {'class':'tableBodyContainer isPL'})
```

```python
data = []

for tr in table.find_all('tr'):
	for td in tr.find_all('td', {'class':'pos button-tooltip'}):
		for span in td.find_all('span', {'class':'value'}):
			position = span.text
			data.append(position)
	for td in tr.find_all('td', {'class':'team'}):
		points = tr.find('td', {'class':'points'}).text
		games = tr.find_all('td')[3].text
		wins = tr.find_all('td')[4].text
		draws = tr.find_all('td')[5].text
		losses = tr.find_all('td')[6].text
		data.extend([games, wins, draws, losses, points])
		for span in td.find_all('span', {'class':'long'}):
			team = span.text
			data.append(team)
```

```python
position = []
team = []
games = []
wins = []
draws = []
losses = []
points = []
for x in range(0, len(data), 7):
    position.append(data[x])
for x in range(6, len(data), 7):
    team.append(data[x])
for x in range(1, len(data), 7):
    games.append(data[x])
for x in range(2, len(data), 7):
    wins.append(data[x])
for x in range(3, len(data), 7):
    draws.append(data[x])
for x in range(4, len(data), 7):
    losses.append(data[x])
for x in range(5, len(data), 7):
    points.append(data[x])
```

```python
data = {'Position':  position,
        'Team': team,
        'Games': games,
        'Wins': wins,
        'Draws': draws,
        'Losses': losses,
        'Points': points,
        }

df = pd.DataFrame (data, columns = ['Position','Team','Games','Wins','Draws','Losses','Points'])
```
